```python
# Version 1: Hardcoded Values

a = 10
b = 20

total_sum = a + b

print("Hardcoded Addition:")
print(a, "+", b, "=", total_sum)
```

```python
# Version 2: Keyboard Input

try:
    x = int(input("Enter first number: "))
    y = int(input("Enter second number: "))

    result = x + y
    print("Addition using keyboard input:", result)

except ValueError:
    print("Please enter only numbers.")
```

```python
# Version 3: Read Single Input from File

# Creating the file
with open("single_input.txt", "w") as f:
    f.write("5")

try:
    with open("single_input.txt", "r") as f:
        num = int(f.read())

    print("Number from file:", num)
    print("Result after multiplying by 2:", num * 2)

except FileNotFoundError:
    print("File not found.")

except ValueError:
    print("Invalid data in file.")
```

```python
# Title: Version 4 – Read From File (Multiple Inputs)
# Save a file with multiple inputs

with open("multiple_inputs.txt", "w") as f:
    f.write("3\n5\n7")

# Reading multiple inputs from a file
try:
    with open("multiple_inputs.txt", "r") as f:
        numbers = [int(line.strip()) for line in f.readlines()]

    result = sum(numbers)
    print(f"Multiple Inputs Read from File: {numbers}, Sum = {result}")

except FileNotFoundError:
    print("File not found!")

except ValueError:
    print("Error in reading or converting the inputs!")
```

```python
# Title: Version 5 – Debugging and Error Handling

try:
    a = int(input("Enter a number: "))
    b = int(input("Enter another number: "))

    result = a / b
    print(f"Division Result: {result}")

except ValueError:
    print("Please enter valid integers!")

except ZeroDivisionError:
    print("Division by zero is not allowed!")
```
